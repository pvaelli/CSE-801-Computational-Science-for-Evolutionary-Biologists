```python
from pandas import *
```

```python
##E matrix for three player game
E=[[0,-1,1],[1,0,-1],[-1,1,0]]
##Beginning frequencies of "Rock", "Paper", and "Scissors"
pa=0.333
pb=0.333
pc=1.0-pa-pb
dt=0.001
ps=[[],[],[]]
for t in range(5000):
    ##Fitness for each choice
    wa=(pa*E[0][0]) + (pb*E[0][1]) + (pc*E[0][2])
    wb=(pa*E[1][0]) + (pb*E[1][1]) + (pc*E[1][2])
    wc=(pa*E[2][0]) + (pb*E[2][1]) + (pc*E[2][2])
    wm=(wa+wb+wc)/3.0
    pa=pa+(dt*(pa*(wa-wm)))
    pb=pb+(dt*(pb*(wb-wm)))
    pc=pc+(dt*(pc*(wc-wm)))
    ##Normalizations
    psum=pa+pb+pc
    pa=(pa/psum)
    pb=(pb/psum)
    pc=(pc/psum)
    ps[0].append(pa)
    ps[1].append(pb)
    ps[2].append(pc)
plot(ps[0],"r")##
plot(ps[1],"b")
plot(ps[2],"g")


```
