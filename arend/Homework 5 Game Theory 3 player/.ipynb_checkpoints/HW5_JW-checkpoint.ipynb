```python
def GT(E,Pa,Pb,Pc,T,dt):
    Ps=[[],[],[]]
    for t in range(T):
        Wa = Pa*E[0][0] + Pb*E[0][1] + Pc*E[0][2]
        Wb = Pb*E[1][0] + Pb*E[1][1] + Pc*E[1][2]
        Wc = Pc*E[2][0] + Pb*E[2][1] + Pc*E[2][2]
        Wm = (Wa + Wb + Wc) / 3.0
        Pa = Pa + dt*Pa*(Wa - Wm)
        Pb = Pb + dt*Pb*(Wb - Wm)
        Pc = Pc + dt*Pc*(Wc - Wm)
        #numerical reasons
        Psum = Pa + Pb + Pc
        Pa = Pa/Psum
        Pb = Pb/Psum
        Pc = Pc/Psum
        Ps[0].append(float(Pa))
        Ps[1].append(float(Pb))
        Ps[2].append(float(Pc))
    return Ps

startPa=[0.1,0.2,0.3,0.35]
startPb=[0.4,0.3,0.6,0.1]
EM = [[0,1,-1],[-1,0,1],[1,-1,0]]

for (Pa,Pb) in zip(startPa,startPb):
    Pc=1.0 - Pa - Pb
    Ps=GT(EM,Pa,Pb,Pc,1000,0.1)
    figure()
    plot(Ps[0], 'r')
    plot(Ps[1], 'g')
    plot(Ps[2], 'b')
    axis([0, 1000, 0, 1])
```
