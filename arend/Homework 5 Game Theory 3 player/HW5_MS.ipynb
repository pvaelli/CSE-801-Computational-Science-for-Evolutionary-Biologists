```python
from pandas import *
```

```python
Pr=0.333
Pp=(1.0-Pr)/2
Psc=(1.0-Pr)/2
E=[[0,-1,+1],[+1,0,-1],[-1,+1,0]]
Ps=[[],[],[]]
dt=0.001
    
for t in range(10000):
    Wr=Pr*E[0][0]+Pp*E[0][1]+Psc*E[0][2]
    Wp=Pr*E[1][0]+Pp*E[1][1]+Psc*E[1][2]
    Wsc=Pr*E[2][0]+Pp*E[2][1]+Psc*E[2][2]
    Wm=(Wr+Wp+Wsc)/3.0
    Pr=Pr+(dt*(Pr*(Wr-Wm)))
    Pp=Pp+(dt*(Pp*(Wp-Wm)))
    Psc=Psc+(dt*(Psc*(Wsc-Wm)))
    Psum=Pr+Pp+Psc
    Pr=Pr/Psum
    Pp=Pp/Psum
    Psc=Psc/Psum
    #numerical reasons
    Ps[0].append(Pr)
    Ps[1].append(Pp)  
    Ps[2].append(Psc)


plot(Ps[0])
plot(Ps[1])
plot(Ps[2])
```
