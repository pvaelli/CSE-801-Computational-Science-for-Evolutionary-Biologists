```python
##adapted from 'Class Notes'
#Create a function to simulate a 'Rock Paper Sissor' like game. 

def GT(Pa,Pb,Pc,T,dt):
    Ps=[[],[],[]]
    E=[[0,-1,1],[1,0,-1],[-1,1,0]]
    for t in range(T):
        Wa=(Pa*E[0][0])+(Pb*E[0][1])+(Pc*E[0][2])
        Wb=(Pa*E[1][0])+(Pb*E[1][1])+(Pc*E[1][2])
        Wc=(Pa*E[2][0])+(Pb*E[2][1])+(Pc*E[2][2])
        Wm=(Wa+Wb+Wc)/3.0
        Pa=Pa+(dt*(Pa*(Wa-Wm)))
        Pb=Pb+(dt*(Pb*(Wb-Wm)))
        Pc=Pc+(dt*(Pc*(Wc-Wm)))
        #numerical reasons
        Psum=Pa+Pb+Pc
        Pa=Pa/Psum
        Pb=Pb/Psum
        Pc=Pc/Psum
        Ps[0].append(float(Pa))
        Ps[1].append(float(Pb))
        Ps[2].append(float(Pc))
    return Ps

Pa=0.33
Pb=(1.0-Pa)/2
Pc=1.0-(Pa+Pb)
Ps=GT(Pa,Pb,Pc,1000,0.1)

plot(Ps[0],'g')
plot(Ps[1],'r')
plot(Ps[2], 'b')

```
