```python
#create a function to simulate rock, paper, scissors
def RPS(Pr,Pp,Ps,T,dt):
    #make a list to keep track of their numbers
    SumList=[[],[],[]]
    #define E, the outcome matrix
    E=[[0.0,-1.0,1.0],[1.0,0.0,-1.0],[-1.0,1.0,0.0]]
    for t in range (T):
        #define W functions, the outcomes of interactions for each species
        Wr=(Pr*E[0][0])+(Pp*E[0][1])+(Ps*E[0][2])
        Wp=(Pr*E[1][0])+(Pp*E[1][1])+(Ps*E[1][2])
        Ws=(Pr*E[2][0])+(Pp*E[2][1])+(Ps*E[2][2])
        Wm=(Wr+Wp+Ws)/3.0
        #define new P's
        Pr=Pr+dt*Pr*(Wr-Wm) # Pr * was missing
        Pp=Pp+dt*Pp*(Wp-Wm) # Pp * was missing
        Ps=Ps+dt*Ps*(Ws-Wm) # Ps * was missing
        #Normalize the new P's
        Ptotal=Pr+Pp+Ps
        Pr=Pr/Ptotal
        Pp=Pp/Ptotal
        Ps=Ps/Ptotal
        #Add the Ps to the list
        SumList[0].append(Pr)
        SumList[1].append(Pp)
        SumList[2].append(Ps)
    return (SumList)

Pr=0.3
Pp=(1.0-Pr)/2.0
Ps=1.0-Pr-Pp
SumList=RPS(Pr,Pp,Ps,10000,0.001)

plot(SumList[0],"r")
plot(SumList[1],"g")
plot(SumList[2],"b")
    
    
```
