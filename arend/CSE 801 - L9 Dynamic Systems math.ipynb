```python
LWstay=[0,0]
LWshift=[0,0]
for n in range(10000):
    doors=[0,0,0]
    doors[random.randint(0,2)]=1
    myChoice=random.randint(0,2)
    doorToOpen=random.randint(0,2)
    while(doors[doorToOpen]==1) or (doorToOpen==myChoice):
        doorToOpen=random.randint(0,2)
    remainingDoor=[0,1,2]
    remainingDoor.remove(myChoice)
    remainingDoor.remove(doorToOpen)
    theOtherDoor=remainingDoor[0]
#    print "m:{0} g:{1} o:{2}".format(myChoice,doorToOpen,theOtherDoor)
    LWstay[doors[myChoice]]=LWstay[doors[myChoice]]+1
    LWshift[doors[theOtherDoor]]=LWshift[doors[theOtherDoor]]+1
print "LW stay: {0}".format(LWstay)
print "LW shift: {0}".format(LWshift)

```

```python
x=10.0
y=10.0
alpha=1.0
beta=0.05
gamma=0.03
delta=0.5
dt=0.1
X=list()
Y=list()
T=850
for t in range(T):
    dx=(alpha*x*dt)-(beta*x*y*dt)
    dy=(gamma*x*y*dt)-(delta*y*dt)
    x=x+dx
    y=y+dy
    Y.append(y)
    X.append(x)
    
plot(X)
plot(Y)
figure()
plot(X,Y)
```
