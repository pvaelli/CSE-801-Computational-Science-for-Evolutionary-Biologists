## Project Description

My project focuses on answering the question of, "which strategy is best for moving through a field of obstacles?" To answer this question, I create a field with obstacles and ask an agent to navigate to the other side. I record their movements including how many times they move straight ahead and how many times they turn. Next, I apply different values to the number of straight movements and turning movements they make corresponding to the time it would take that agent to either move straight forward or turn. In this project, Dori will be better at moving forward (faster, take a shorter amount of time to move forward) and Marlin will be better at turning (more agile, take a shorter amount of time to turn). By altering exactly how much faster each agent is at either moving forward (Dori) or turning (Marlin) or the number of obstacles present (a total of three independent variables), I can measure in which situations each individual would be better.

My hypothesis is that if you increase the number of obstacles, the more agile agent will become quicker than the agent who is faster at moving straight forward.

This model puts boundaries on three edges of the map and leaves the "finish line" open because (theoretically) when the agent reaches the "finish line," the simulation will stop. Agents only have the option to move in four cardinal directions and cannot move diagonally. This is to make the simulation more simple and much faster. The model is also limited by the method by which the agents move through the space. This movement is purposeful, meaning that the agents "know" where they have to go and will never move away from the goal. A gradient is created from the finish line moving towards the starting line. This serves two purposes: first, to ensure an open path exists to the finish line and second, to guide the agent because the agent will only move to a space that has a gradient value of less or equal value compared to the space it is currently on. Therefore, the agent has no possibility to wander. This helps with the speed of the simulation, but also gives the model some specificity because in most biological races through space, the agent knows where it wants to go and will make purposeful movements towards that goal.

The analysis of the data includes measuring the average time it takes each agent to move through a space with a given set of parameters (number of obstacles, time to move straight forward, and time to turn) then take the difference of the two strategies (Dori and Marlin). This difference will tell us who wins the race and can be plotted on heat maps to show how the success changes when independent variables are manipulated.

```python
# Here is the code I used to create the field and give it gradient values.
# This code is annotated but is later converted into an object oriented format to make the large scale runs easier.

#Create a grid which is a square "sizeOfGrid" long and "sizeOfGrid" wide
def createGrid(sizeOfGrid,grid):
    #The (+2) is to account for the two walls on either side
    for i in range(sizeOfGrid+2):#
        grid.append([])
        #The (+1) is to account for the wall at the base of the grid
        for j in range(sizeOfGrid+1):#
            #label the open spaces "0"
            grid[i].append(0)
    return grid

#Place some obstacles in the grid...
def placeObstacles(numberOfObstacles,sizeOfGrid,grid):
    tempListX=list()
    tempListY=list()
    #Create lists of all of the possible x and y values an obstacle could have
    for i in range(sizeOfGrid):
        tempListX.append(i+1)
        tempListY.append(i+1)
    for n in range(numberOfObstacles):
        #shuffle the lists and take the first pair of points as the possibility for the position of the obstacle
        random.shuffle(tempListX)
        random.shuffle(tempListY)
        x=0
        y=0
        xPosition=tempListX[x]
        yPosition=tempListY[y]
        #But make sure there wasn't already one where you try to put the next one
        while grid[xPosition][yPosition]==1:
            try:
                #if there was already an obstacle there, try the next pair of points
                while grid[xPosition][yPosition]==1:
                    x=x+1
                    xPosition=tempListX[x]
                    y=y+1
                    yPosition=tempListY[y]
            except IndexError:
                #if you've exhausted the current list of points to try, create a new set of pairs of points to try
                random.shuffle(tempListX)
                random.shuffle(tempListY)
                x=0
                y=0
                xPosition=tempListX[x]
                yPosition=tempListY[y]
        #Label the obstacles "1"
        grid[xPosition][yPosition]=1
    return grid

def rank(sizeOfGrid,grid):
    labels=[]
    #This "maxValue" will serve as an extremely high value in the grid that is impossible to achieve by the normal gradient
    #Therefore, by placing this value, or any value above it, it will serve as a barrier to movement
    maxValue=sizeOfGrid**2
    #create a list for later...
    reverseDirection=list()
    for l in range(sizeOfGrid):
        reverseDirection.append(sizeOfGrid-l)#
    #create a grid of zeros to keep track of the labels
    for i in range(sizeOfGrid+2):#
        labels.append([])
        for j in range(sizeOfGrid+1):#
            labels[i].append(0)
            #label obstacles with this maxValue
            if grid[i][j]==1:
                labels[i][j]=maxValue
            #create borders except for the last y row by labeling them with this maxValue
            if i==0 or i==(sizeOfGrid+1) or j==0: #
                labels[i][j]=maxValue #
    #for the last row, if the space is empty, place a 1, then work backwards
    for j in range(sizeOfGrid):
        if grid[j+1][sizeOfGrid]==0: #
            labels[j+1][sizeOfGrid]=1 #
    #working backwards...k is the y coordinate, increase the value each time you move backwards, creating a gradient
    for k in range(sizeOfGrid-1):
        #m is the x coordinate, if there is a number straight in front of it, add one and place a label
        for m in range(sizeOfGrid):
            if labels[m+1][sizeOfGrid-1-k]==0 and labels[m+1][sizeOfGrid-k]<maxValue:#
                labels[m+1][sizeOfGrid-1-k]=labels[m+1][sizeOfGrid-k]+1#
        #n is the x coordinate, work from left to right and if there are numbers to the side, add one to the smaller number and place a label
        for n in range(sizeOfGrid):
            if labels[n+1][sizeOfGrid-1-k]==0: #
                #if there is an obstacle ahead and you are in the middle of the grid...
                if labels[n+1][sizeOfGrid-k]>=maxValue and 1<n<(sizeOfGrid):
                    #take the smaller one given that it's not a zero
                    tempList=list()
                    tempList.append(labels[n+2][sizeOfGrid-1-k])#
                    tempList.append(labels[n][sizeOfGrid-1-k])#
                    try:
                        tempList.remove(0)
                        tempList.remove(0)
                        labels[n+1][sizeOfGrid-1-k]=0#
                    except ValueError:
                        labels[n+1][sizeOfGrid-1-k]=min(tempList)+1#
                #if there is an obstacle ahead and the position is up against the right wall, automatically take the number to your left
                if labels[n+1][sizeOfGrid-k]>=maxValue and n==sizeOfGrid:#
                    labels[n+1][sizeOfGrid-1-k]=labels[n][sizeOfGrid-1-k]+1#
        #p is the x coordinate, work from right to left and if there are numbers to the side, add one to the smaller number and place a label
        for p in reverseDirection:
            if labels[p][sizeOfGrid-1-k]==0:#
                if labels[p][sizeOfGrid-k]>=maxValue and 1<p<(sizeOfGrid):#
                    #take the smaller one given that it's not a zero
                    tempList=list()
                    tempList.append(labels[p+1][sizeOfGrid-1-k])#
                    tempList.append(labels[p-1][sizeOfGrid-1-k])#
                    try:
                        tempList.remove(0)
                        tempList.remove(0)
                        labels[p][sizeOfGrid-1-k]=0#
                    except ValueError:
                        labels[p][sizeOfGrid-1-k]=min(tempList)+1#
                #if there is an obstacle ahead and the position is up against the left wall, automatically take the number to your right
                if labels[p][sizeOfGrid-k]>=maxValue and p==1:#
                    labels[p][sizeOfGrid-1-k]=labels[p+1][sizeOfGrid-1-k]+1#
    return labels

sizeOfOcean=10
ocean=[]
numberOfObstacles=25
maxValue=sizeOfOcean**2

ocean=createGrid(sizeOfOcean,ocean)
ocean=placeObstacles(numberOfObstacles,sizeOfOcean,ocean)
rankList=rank(sizeOfOcean,ocean)
#print rankList
#This shows the grid with only obstacles
imshow(ocean,cmap='bwr',interpolation="nearest")
figure()
#This shows the grid with obstacles, borders, and removes paths where it is impossible to get to the finish line
imshow(rankList,cmap='bwr',interpolation="nearest")

#Because of the maxValues, some of the values in the original "rankList" are obscured
#therefore, if I remove these large values, you can visualize the path the agent will take
rankListPrime=list()
for i in range(sizeOfOcean+2):
    rankListPrime.append([])
    for j in range(sizeOfOcean+1):
        if rankList[i][j]>=maxValue:
            rankListPrime[i].append(0)
        else:
            rankListPrime[i].append(rankList[i][j])

figure()
imshow(rankListPrime,cmap='bwr',interpolation="nearest")
print rankList
```

```python
#####This is the above functions put into a class system with the added functions of checking the grid for a viable path,
#place the fish, move the fish, and apply the time-of-movement parameters to the number of movements and turns the
#agent made.

class Fish:
    def __init__(self,sizeOfGrid):
        self.trackX=[]
        self.trackY=[]
        maxValue=sizeOfGrid**2
        self.time=0
    
    def makeGrid(self,sizeOfGrid,numberOfObstacles):
        #create grid
        self.grid=[]
        for i in range(sizeOfGrid+2):
            self.grid.append([])
            for j in range(sizeOfGrid+1):
                #label the open spaces "0"
                self.grid[i].append(0)
        #place obstacles
        tempListX=list()
        tempListY=list()
        for d in range(sizeOfGrid):
            tempListX.append(d+1)
            tempListY.append(d+1)
        for n in range(numberOfObstacles):
            random.shuffle(tempListX)
            random.shuffle(tempListY)
            x=0
            y=0
            xPosition=tempListX[x]
            yPosition=tempListY[y]
            #But make sure there wasn't already one where you try to put the next one
            while self.grid[xPosition][yPosition]==1:
                try:
                    while self.grid[xPosition][yPosition]==1:
                        x=x+1
                        xPosition=tempListX[x]
                        y=y+1
                        yPosition=tempListY[y]
                except IndexError:
                    random.shuffle(tempListX)
                    random.shuffle(tempListY)
                    x=0
                    y=0
                    xPosition=tempListX[x]
                    yPosition=tempListY[y]
            #Label the obstacles "1"
            self.grid[xPosition][yPosition]=1
        
    def createRanks(self,sizeOfGrid):
        #create a list for later...
        self.reverseDirection=list()
        for l in range(sizeOfGrid):
            self.reverseDirection.append(sizeOfGrid-l)
        #create a grid of zeros to keep track of the labels
        self.labels=[]
        for i in range(sizeOfGrid+2):
            self.labels.append([])
            for j in range(sizeOfGrid+1):
                self.labels[i].append(0)
                if self.grid[i][j]==1:
                    self.labels[i][j]=maxValue
                if i==0 or i==(sizeOfGrid+1) or j==0:
                    self.labels[i][j]=maxValue
        #for the last row, if the space is empty, place a 1, then work backwards
        for j in range(sizeOfGrid):
            if self.grid[j+1][sizeOfGrid]==0:
                self.labels[j+1][sizeOfGrid]=1
        #working backwards...k is the y coordinate
        for k in range(sizeOfGrid-1):
            #m is the x coordinate, if there is a number straight in front of it, add one and place a label
            for m in range(sizeOfGrid):
                if self.labels[m+1][sizeOfGrid-1-k]==0 and self.labels[m+1][sizeOfGrid-k]<maxValue:
                    self.labels[m+1][sizeOfGrid-1-k]=self.labels[m+1][sizeOfGrid-k]+1
            #n is the x coordinate, work from left to right and if there are numbers to the side, add one to the smaller number and place a label
            for n in range(sizeOfGrid):
                if self.labels[n+1][sizeOfGrid-1-k]==0:
                    #if there is an obstacle ahead and you are in the middle of the grid...
                    if self.labels[n+1][sizeOfGrid-k]>=maxValue and 1<n<(sizeOfGrid):
                        #take the smaller one given that it's not a zero
                        tempList=list()
                        tempList.append(self.labels[n+2][sizeOfGrid-1-k])
                        tempList.append(self.labels[n][sizeOfGrid-1-k])
                        try:
                            tempList.remove(0)
                            tempList.remove(0)
                            self.labels[n+1][sizeOfGrid-1-k]=0
                        except ValueError:
                            self.labels[n+1][sizeOfGrid-1-k]=min(tempList)+1
                    #if there is an obstacle ahead and the position is up against the right wall, automatically take the number to your left
                    if self.labels[n+1][sizeOfGrid-k]>=maxValue and n==sizeOfGrid:
                        self.labels[n+1][sizeOfGrid-1-k]=self.labels[n][sizeOfGrid-1-k]+1
            #p is the x coordinate, work from right to left and if there are numbers to the side, add one to the smaller number and place a label
            for p in self.reverseDirection:
                if self.labels[p][sizeOfGrid-1-k]==0:
                    if self.labels[p][sizeOfGrid-k]>=maxValue and 1<p<(sizeOfGrid):
                        #take the smaller one given that it's not a zero
                        tempList=list()
                        tempList.append(self.labels[p+1][sizeOfGrid-1-k])
                        tempList.append(self.labels[p-1][sizeOfGrid-1-k])
                        try:
                            tempList.remove(0)
                            tempList.remove(0)
                            self.labels[p][sizeOfGrid-1-k]=0
                        except ValueError:
                            self.labels[p][sizeOfGrid-1-k]=min(tempList)+1
                    #if there is an obstacle ahead and the position is up against the left wall, automatically take the number to your right
                    if self.labels[p][sizeOfGrid-k]>=maxValue and p==1:
                        self.labels[p][sizeOfGrid-1-k]=self.labels[p+1][sizeOfGrid-1-k]+1
    
    def placeFish(self,sizeOfGrid):
        #put it in the first row but only in a place where it has the possibility of making it to the end
        self.trackY.append(1)
        newX=random.randint(1,sizeOfGrid)
        while self.labels[newX][1]>=(sizeOfGrid**2) or self.labels[newX][1]<sizeOfGrid:
            newX=random.randint(1,sizeOfGrid)
            #print "used place"
        self.trackX.append(newX)
    
    def moveFish(self,sizeOfGrid):
        #a movement matrix
        xMove=[0,0,1,-1]
        yMove=[1,-1,0,0]
        #this list will be shuffled then the first number in the list will be used as an index in the movement matrix
        tempList=[0,1,2,3]
        for i in range(4):
            tempList.append(tempList.pop(random.randint(0,3)))
        while (self.labels[xMove[tempList[0]]+self.trackX[len(self.trackX)-1]][yMove[tempList[0]]+self.trackY[len(self.trackY)-1]])>(self.labels[self.trackX[len(self.trackX)-1]][self.trackY[len(self.trackY)-1]]):
            #if that movement is not possible, remove that movement index as a possibility and try again
            #situations where the agent will get stuck have been labeled as barriers, so this should not fail
            tempList.pop(0)
        self.trackX.append(xMove[tempList[0]]+self.trackX[len(self.trackX)-1])
        self.trackY.append(yMove[tempList[0]]+self.trackY[len(self.trackY)-1])
    
    def checkGrid(self,sizeOfGrid,numberOfObstacles):
        #compile the labels in the first row where the fish will be placed
        self.xValues=list()
        for k in range(sizeOfGrid):
            self.xValues.append(self.labels[k+1][1])
        #if there is no possible placement for the fish, try again
        while min(self.xValues)>=maxValue:
            self.xValues=list()
            self.makeGrid(sizeOfGrid,numberOfObstacles)
            self.createRanks(sizeOfGrid)
            for k in range(sizeOfGrid):
                self.xValues.append(self.labels[k+1][1])
            #print "used check"
                
    def calculateTime(self,turn,straight):
        self.xDiff=[]
        self.yDiff=[]
        #calculate the change in each direction
        for i in range(len(self.trackX)-1):
            self.xDiff.append(abs(self.trackX[i+1]-self.trackX[i]))
            self.yDiff.append(abs(self.trackY[i+1]-self.trackY[i]))
        for j in range(len(self.xDiff)-1):
            #if there is a change in direction, add the "time to turn" to the total time, x should be the same as y so only do this once
            if self.xDiff[j+1]!=self.xDiff[j]:
                self.time=self.time+turn
        for k in range(len(self.xDiff)):
            #if there was a movement in that direction, add the "time to go straight" to the total time
            if self.xDiff[k]==1:
                self.time=self.time+straight
            if self.yDiff[k]==1:
                self.time=self.time+straight
    
    def __str__(self):
        return "X:{0} Y:{1}".format(self.trackX[len(self.trackX)-1],self.trackY[len(self.trackY)-1])

sizeOfOcean=100
numberOfJellyfish=40
maxValue=sizeOfOcean**2

#Here are a list of commands that will get the fish to the finish line. This is later put into a single function

dori=Fish(sizeOfOcean)

dori.makeGrid(sizeOfOcean,numberOfJellyfish)
dori.createRanks(sizeOfOcean)
dori.checkGrid(sizeOfOcean,numberOfJellyfish)
dori.placeFish(sizeOfOcean)

#print dori

imshow(dori.grid,cmap='bwr',interpolation="nearest")
figure()
imshow(dori.labels,cmap='bwr',interpolation="nearest")

while dori.trackY[len(dori.trackY)-1]<(sizeOfOcean):
    dori.moveFish(sizeOfOcean)
    #print dori

dori.calculateTime(0.785,1)
print dori.time
```

In early tests when I only had one independent variable (number of obstacles), my hypothesis seemed to be rejected and Dori seemed to always win. 

At this point I employed a linear regression of the "differenceBwMeans" to to see if the speed of the two agents converges at all. If there is some convergence, then it could mean that I need to test more obstacles or more values of movement. The code for this linear regression is used below in a much more expansive format but originally, it was used here to determine if Marlin has any chance at over-taking Dori.

When I saw some convergence, I decided to add the additional independent variables of the exact times of movement. Originally, the normal values were "one second" for forward movement and "0.785 seconds" to turn. This is based on the distance to move 90 degrees circularly within a square of side length one. Circumference = diameter*pi, diameter=1 so C=1*pi, C/4 = pi/4 = 0.785. Also, originally, Dori was considered to be twice as fast at moving forward (forward:0.5, turn:0.785) and Marlin was twice as fast at turning (forward:1, turn:0.3925). When I saw that Marlin was never winning, I expanded the range of "how much faster" each of these agents were to see "how fast an agent needs to be in order to win."

These expanded independent variables are shown below.

```python
##### This is the primary set of code that will test all three independent variables at the same time
#This code can take hours/days to achieve an appropriate number of trials. For quick testing, try trial sizes of "2"
sizeOfOcean=100
trials=2
numbersToTestObstacles=[]
numbersToTestSpeed=[]
numbersToTestForwardSpeed=[]
collectionOfMeansDori=[]
collectionOfMeansMarlin=[]
collectionOfSEMDori=[]
collectionOfSEMMarlin=[]
differenceBwMeans=list()
c=0


for f in range(int(((sizeOfOcean**2)/2)/250)):
    numbersToTestObstacles.append(f*250+1000)
for j in range(int((.785-.1)/.1)):
    numbersToTestSpeed.append((j*.1)+.1)
for a in range(int((1-0.5)/0.1)):
    numbersToTestForwardSpeed.append((a*0.1)+0.5)
for k in numbersToTestSpeed:
    print k
    collectionOfMeansDori.append([])
    collectionOfMeansMarlin.append([])
    collectionOfSEMDori.append([])
    collectionOfSEMMarlin.append([])
    differenceBwMeans.append([])
    h=0
    for b in numbersToTestForwardSpeed:
        print b
        collectionOfMeansDori[c].append([])
        collectionOfMeansMarlin[c].append([])
        collectionOfSEMDori[c].append([])
        collectionOfSEMMarlin[c].append([])
        for g in numbersToTestObstacles:
            doriRecord=[]
            marlinRecord=[]
            for i in range(trials):
                marlin=Fish(sizeOfOcean)
                dori=Fish(sizeOfOcean)
                doriRecord.append(testFish(dori,sizeOfOcean,g,0.785,b))
                marlinRecord.append(testFish(marlin,sizeOfOcean,g,k,1))
            print g
            collectionOfMeansDori[c][h].append(int(mean(doriRecord)))
            collectionOfMeansMarlin[c][h].append(int(mean(marlinRecord)))
            collectionOfSEMDori[c][h].append(std(doriRecord))
            collectionOfSEMMarlin[c][h].append(std(marlinRecord))
        differenceBwMeans[c].append([])
        for u in range(len(collectionOfMeansDori[c][h])):
            differenceBwMeans[c][h].append(collectionOfMeansDori[c][h][u]-collectionOfMeansMarlin[c][h][u])
        h=h+1
    c=c+1
```

These images show that in order for Marlin to win while Dori is twice as fast at moving forward, Marlin needs to be almost eight times as fast at turning (Marlin time to turn=0.1 while Dori time forward =0.5, graph 1). Therefore, it is hard to draw conclusions from the heat maps if an agent becomes more successful in and of itself, but when compared to the other agent, the conclusions are clear. In some of the early plots (graphs 1 and 2), you can see that as the number of obstacles increase, more red appears. This supports my hypothesis but not to the drastic degree I expected. Ultimately, I can conclude it is better to be faster at moving forward than it is to be better at turning.
