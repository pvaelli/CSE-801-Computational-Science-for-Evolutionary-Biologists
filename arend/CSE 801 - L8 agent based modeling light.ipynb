```python
import random

def countTheZoo(theArea):
    myCounter=[0,0,0]
    for x in range(len(theArea)):
        for y in range(len(theArea[x])):
            myCounter[theArea[x][y]]=myCounter[theArea[x][y]]+1                
    return myCounter

def populateZoo(theArea,withWhat,howMany):
    for n in range(howMany):
        x=random.randint(0,xDim-1)
        y=random.randint(0,yDim-1)
        while area[x][y]!=0:
            x=random.randint(0,xDim-1)
            y=random.randint(0,yDim-1)
        theArea[x][y]=withWhat #add withWhat 
    return theArea

def getNewLocation(x,y,xBound,yBound):
    xO=0
    yO=0
    while ((xO==0)and(yO==0)) or ((x+xO)<0) or ((x+xO)>=xBound) or ((y+yO)<0) or ((yO+y)>=yBound):
        xO=random.randint(-1,1)
        yO=random.randint(-1,1)        
    return x+xO,y+yO

#initialization
area=list()
xDim=50
yDim=50
nStartBunny=88
nStartRaptor=123

#make area empty
for x in range(xDim):
    area.append(list())
    for y in range(yDim):
        area[x].append(0)
        
area=populateZoo(area,1,nStartBunny)
area=populateZoo(area,2,nStartRaptor)
            
imshow(area,interpolation="nearest")
print "our function" +str(countTheZoo(area))
bunny=list()
raptor=list()
#main simulation loop
countCounter=0
for t in range(1000000):
    countCounter=countCounter+1
    #pick random location
    x=random.randint(0,xDim-1)
    y=random.randint(0,yDim-1)
    
    if area[x][y]==1: #bunny
        #do bunny thing
        nX,nY=getNewLocation(x,y,xDim,yDim)
        
        if random.random()>0.5:
            #do the hopping
            if area[nX][nY]==0:
                area[nX][nY]=1
                area[x][y]=0
        else:
            #do offspring
            if area[nX][nY]==0:
                area[nX][nY]=1
    else:
        if area[x][y]==2:
            nX,nY=getNewLocation(x,y,xDim,yDim)
            if area[nX][nY]==1:
                #jumped on a bunny -> offspring
                area[nX][nY]=2
            else:
                if area[nX][nY]==0:
                    #jumped on an empty spot -> just jump
                    area[nX][nY]=2
                    area[x][y]=0
    #remove the living
    if random.random()<0.1:
        x=random.randint(0,xDim-1)
        y=random.randint(0,yDim-1)
        area[x][y]=0
    if countCounter>1000:
        V=countTheZoo(area)
        bunny.append(V[1])
        raptor.append(V[2])
        countCounter=0
figure()
plot(bunny,'g')
plot(raptor,'r')
```
