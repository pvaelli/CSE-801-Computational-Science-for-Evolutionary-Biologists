```python
aveY=[0.0]*1000

for i in range(10):
    data=read_csv("Data_{0}.csv".format(i))
    for j in range(len(data["T"])-1):
        #print "{0} {1}".format(data["T"][j],data["T"][j+1])
        start=data["T"][j]
        end=data["T"][j+1]
        for k in range(start,end):
            aveY[k]=aveY[k]+data["Y"][j]
    for k in range(end,1000):
        aveY[k]=aveY[k]+data["Y"][len(data["Y"])-1]

for j in range(1000):
    aveY[j]=float(aveY[j])/float(10)
plot(range(1000),aveY)
```

```python
myUrl="http://www.fullbooks.com/Second-April.html"
theSite=str(urllib.urlopen(myUrl).read())
```

```python

theSite=theSite.replace("<br>","")
theSite=theSite.replace("\n"," ")
theSite=theSite.replace("\r","")
theSite=theSite.replace("\t","")
theSite=theSite.replace("?","")
theSite=theSite.replace("!","")
theSite=theSite.replace(",","")
theSite=theSite.replace(".","")
words=theSite.split(" ")
```

```python
D=dict()
for w in words:
    if w in D:
        D[w]=D[w]+1
    else:
        D[w]=1
```

```python
def fibo(n):
    if n==0:
        return 0
    if n==1:
        return 1
    return fibo(n-1)+fibo(n-2)

print fibo(19)
```
