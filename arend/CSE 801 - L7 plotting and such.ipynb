```python
def swap(A,B):
    return B,A

N=1000
L=[0]*(N/2)+[1]*(N/2)
Collector=list()
for i in range(N*10):
    x1=random.randint(0,N-1)
    x2=random.randint(0,N-1)
    while(x1==x2):
        x1=random.randint(0,N-1)
        x2=random.randint(0,N-1)
    L[x1],L[x2]=swap(L[x1],L[x2])
    left=0
    right=0
    for j in range(N/2):
        left=left+L[j]
        right=right+L[(N/2)+j]
    Collector.append(right-left)
plot(Collector)
    #Collector.append(list(L))
#imshow(Collector,interpolation='nearest')
```
