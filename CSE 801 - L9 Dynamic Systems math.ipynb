```python
# Mathematical formulas for Lotka Volterra dynamics

predator=10.0
prey=10.0
alpha=1.0
beta=0.05
gamma=0.02
delta=0.5
dt=0.1

X=list()
Y=list()
T=1000
for t in range(T):
    #dx=(prey*(alpha - predator*beta))*dt
    #dy=(-predator*(gamma - delta*prey))*dt
    dx=(alpha*prey*dt)-(beta*prey*predator*dt)
    dy=(gamma*prey*predator*dt)-(delta*predator*dt)
    predator=predator+dy
    prey=prey+dx
    X.append(prey)
    Y.append(predator)

plot(X)
plot(Y)
figure()
plot(X,Y)
```
