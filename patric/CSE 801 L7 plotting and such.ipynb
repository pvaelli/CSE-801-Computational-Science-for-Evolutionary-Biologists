```python
L=[0]*(N/2) +[1]*(N/2)

def swap(A,B):
    return B,A

Collector=list()
for i in range(100):
    x1=random.randint(0,N-1)
    x2=random.randint(0,N-1)
    while(x1==x2):
        x1=random.randint(0,N-1)
        x2=random.randint(0,N-1)
    L[x1],L[x2]=swap(L[x1],L[x2])
    Collector.append(list(L))
    left=0
    for j in range(N/2):
        left=left+L[i]
    right=0
    for j in range(N/2, N/2):
        right=right+L[j]
    Collector.append(right-left)
plot(Collector)
```
